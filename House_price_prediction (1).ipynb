## Feature engineering

```python
df.head()
```

```python
df= df.drop(['id'], axis=1)
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year']= df['date'].apply(lambda date: date.year)
df['month']= df['date'].apply(lambda date: date.month)
```

```python
df.head()
```

```python
sns.boxplot(x='month', y='price', data=df)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
df= df.drop('date', axis=1)
```

```python
df.head()
```

```python
# zipcode is a tricky column when feeding into model, as its a numerical value, the model might assume that one zipcode is simply better than the other, which does not make sense unless the zipcode is completely in line with the lat and long data
```

```python
df['zipcode'].value_counts()
```

```python
#70 categories is too much so we will drop this
#with domain experience one can evaluate this column and group them on lat and long
```

```python
df= df.drop('zipcode', axis=1)
```

```python
df['yr_renovated'].value_counts()
```

```python
# most houses show 0 which means they were not renovated, therefore we can change this column to indicate whether the house was renovated or not 
```

```python
df['sqft_basement'].value_counts()
```

```python
X= df.drop('price', axis=1).values
y= df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
import tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model= Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train,
          validation_data=(X_test, y_test),
          batch_size=128,
         epochs=400)
```

```python
losses= pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,8))
losses.plot()
```

```python
# if the orange line went up towards the end of the graph that indicates that the model is overfit, however in our case we can see that the losses remain constant after a point
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
prediction= model.predict(X_test)
```

```python
prediction
```

```python
np.sqrt(mean_squared_error(y_test, prediction))
```

```python
mean_absolute_error(y_test, prediction)
```

### whether this value of mean absolute error is acceptable or not? To verify this we can take a look at the mean value of the prices in our dataset

```python
df.describe()
```

### As we can see here that the mean value of prices is around 500000 dollars, therefore we are off by about 20% with our model, this value is neither very good nor very bad. 

```python
explained_variance_score(y_test,prediction)
```

### This value of varaiance is decent. however, whether this model is good or not depends on whether or not there exists a similar model which is able to produce better results.

```python
plt.figure(figsize=(12,8))
plt.scatter(y_test, prediction)
plt.plot(y_test,y_test,'r')
```

### We can probably retrain the model on the bottom 99% of the data
